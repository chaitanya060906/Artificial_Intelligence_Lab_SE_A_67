```python
import numpy as np

# Define matrix
A = np.array([[1, 2, 3],
              [0, 1, 4],
              [5, 6, 0]])

# Compute determinant
det_A = np.linalg.det(A)

print("Matrix A:")
print(A)
print("\nDeterminant of A:")
print(det_A)
```

```python
# Compute inverse
A_inv = np.linalg.inv(A)

print("\nInverse of A:")
print(A_inv)
```

```python
# Coefficient matrix
B = np.array([[2, 3],
              [4, 5]])

# Constants vector
C = np.array([10, 20])

# Solve system
solution = np.linalg.solve(B, C)

print("\nSolution (x, y):")
print(solution)
```
