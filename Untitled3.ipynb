```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_wine

# ------------------------------
# Load Dataset
# ------------------------------
wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df["Class"] = wine.target

print("\n=========== WINE DATASET ANALYSIS ===========\n")

# ------------------------------
# Show basic information
# ------------------------------
print("Total Number of Records:", df.shape[0])
print("Total Number of Features:", df.shape[1])
print("\n")

# ------------------------------
# Focus on one attribute: alcohol
# ------------------------------
print("----- Alcohol Attribute Analysis -----")
print("Minimum Alcohol Value :", round(df["alcohol"].min(), 2))
print("Maximum Alcohol Value :", round(df["alcohol"].max(), 2))
print("Average Alcohol Value :", round(df["alcohol"].mean(), 2))
print("\n")

# ------------------------------
# Feature-wise Aggregate Summary
# ------------------------------
print("----- Feature-wise Summary (Mean Values) -----\n")

for column in df.columns:
    if column != "Class":
        print(f"{column} → Mean: {round(df[column].mean(), 2)}")

print("\n")

# ------------------------------
# Group by Class
# ------------------------------
print("----- Class-wise Alcohol Average -----\n")

class_group = df.groupby("Class")["alcohol"].mean()

for cls, value in class_group.items():
    print(f"Class {cls} → Average Alcohol: {round(value, 2)}")

print("\n")

# ------------------------------
# Count of each Class
# ------------------------------
print("----- Number of Wines in Each Class -----\n")

counts = df["Class"].value_counts()

for cls, count in counts.items():
    print(f"Class {cls} has {count} wines")

print("\n=========== ANALYSIS COMPLETED ===========")
```
