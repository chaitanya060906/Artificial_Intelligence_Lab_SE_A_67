### 1. Program to compute the factorial of a given number

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

# Example usage:
num = 5
print(f"The factorial of {num} is {factorial(num)}")

num = 0
print(f"The factorial of {num} is {factorial(num)}")

num = -3
print(f"The factorial of {num} is {factorial(num)}")
```

### 2. Program to display the multiplication table for a given number

```python
def display_table(n, limit=10):
    print(f"Multiplication Table for {n}:")
    for i in range(1, limit + 1):
        print(f"{n} x {i} = {n * i}")

# Example usage:
num = 7
display_table(num)

print("\n")
num = 12
display_table(num, limit=5)
```

### 3. Program to identify if a given number is odd or even

```python
def check_odd_even(n):
    if n % 2 == 0:
        return f"{n} is an even number."
    else:
        return f"{n} is an odd number."

# Example usage:
num = 4
print(check_odd_even(num))

num = 7
print(check_odd_even(num))

num = 0
print(check_odd_even(num))
```

### 4. Program to identify if the entered number is a palindrome or not

```python
def is_palindrome(n):
    # Convert number to string to easily reverse it
    s = str(n)
    if s == s[::-1]:
        return f"{n} is a palindrome number."
    else:
        return f"{n} is not a palindrome number."

# Example usage:
num = 121
print(is_palindrome(num))

num = 123
print(is_palindrome(num))

num = 1221
print(is_palindrome(num))
```

### 5. Develop an Expert System that provides simple decision-making

An Expert System typically consists of a knowledge base (facts and rules) and an inference engine (to apply the rules to make decisions). Below is a very simple example of a rule-based expert system for suggesting an activity based on the weather and time of day.

```python
class SimpleExpertSystem:
    def __init__(self):
        self.knowledge_base = [
            {"condition": {"weather": "sunny", "time": "day"}, "action": "Go for a walk in the park."},
            {"condition": {"weather": "rainy", "time": "day"}, "action": "Read a book indoors."},
            {"condition": {"weather": "sunny", "time": "night"}, "action": "Stargaze or enjoy the night sky."},
            {"condition": {"weather": "rainy", "time": "night"}, "action": "Watch a movie or play board games."},
            {"condition": {"weather": "cloudy", "time": "day"}, "action": "Visit a museum or art gallery."}
        ]

    def recommend_activity(self, weather, time):
        for rule in self.knowledge_base:
            conditions_met = True
            for key, value in rule["condition"].items():
                if key == "weather" and weather != value:
                    conditions_met = False
                    break
                if key == "time" and time != value:
                    conditions_met = False
                    break
            if conditions_met:
                return rule["action"]
        return "Sorry, I don't have a recommendation for these conditions."

# Example usage of the Expert System:
es = SimpleExpertSystem()

weather_input = "sunny"
time_input = "day"
print(f"Given weather: {weather_input}, time: {time_input} -> Recommendation: {es.recommend_activity(weather_input, time_input)}")

weather_input = "rainy"
time_input = "night"
print(f"Given weather: {weather_input}, time: {time_input} -> Recommendation: {es.recommend_activity(weather_input, time_input)}")

weather_input = "snowy"
time_input = "day"
print(f"Given weather: {weather_input}, time: {time_input} -> Recommendation: {es.recommend_activity(weather_input, time_input)}")
```
