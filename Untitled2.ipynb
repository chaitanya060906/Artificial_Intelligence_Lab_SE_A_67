```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate normal distribution data
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=10000)

# Plot histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', label='Histogram')

# Plot theoretical normal distribution curve
x = np.linspace(min(data), max(data), 1000)
plt.plot(x, norm.pdf(x, 0, 1), 'r', linewidth=2, label='Normal PDF')

plt.title("Normal Distribution Histogram")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
mean = np.mean(data)
median = np.median(data)
variance = np.var(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
# Simulate coin toss (0 = Tails, 1 = Heads)
tosses = np.random.choice([0, 1], size=10000)

# Count outcomes
heads = np.sum(tosses == 1)
tails = np.sum(tosses == 0)

# Compute probabilities
prob_heads = heads / len(tosses)
prob_tails = tails / len(tosses)

print("Number of Heads:", heads)
print("Number of Tails:", tails)
print("Probability of Heads:", prob_heads)
print("Probability of Tails:", prob_tails)
```

```python
plt.bar(['Heads', 'Tails'], [prob_heads, prob_tails], color=['blue', 'green'])
plt.title("Coin Toss Probability")
plt.ylabel("Probability")
plt.show()
```
